```python

import ir_datasets

# Load the dataset
dataset = ir_datasets.load("neuclir/1/ru")

# Display a few documents to understand the structure
for i, doc in enumerate(dataset.docs_iter()):
    print(f"Document ID: {doc.doc_id}")
    print(f"Title: {doc.title}")
    print(f"Text: {doc.text}")
    print(f"URL: {doc.url}")
    print(f"Time: {doc.time}")
    print(f"CC File: {doc.cc_file}")
    print("-" * 50)
    if i == 4:  # Limit to the first 5 documents
        break
```

```python

from googletrans import Translator
import pandas as pd

# Initialize the translator
translator = Translator()

# Create a list to store translated documents
translated_docs = []

# Iterate through the dataset
for i, doc in enumerate(dataset.docs_iter()):
    # Translate title and text
    translated_title = translator.translate(doc.title, src='ru', dest='en').text if doc.title else None
    translated_text = translator.translate(doc.text, src='ru', dest='en').text if doc.text else None
    
    # Add translated document to the list
    translated_docs.append({
        "doc_id": doc.doc_id,
        "original_title": doc.title,
        "translated_title": translated_title,
        "original_text": doc.text,
        "translated_text": translated_text,
        "url": doc.url,
        "time": doc.time,
        "cc_file": doc.cc_file
    })
    
    # Break after a few documents for testing purposes
    if i == 10:  # Adjust this as needed
        break

# Convert the list to a DataFrame
translated_df = pd.DataFrame(translated_docs)

# Save to Excel for inspection
translated_df.to_excel("translated_documents.xlsx", index=False)

# Display a sample of the translated DataFrame
translated_df.head()


```
