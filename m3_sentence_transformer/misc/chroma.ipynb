```python
from FlagEmbedding import BGEM3FlagModel
from tqdm.notebook import tqdm
from m3_sentence_transformer.misc.preprocessing import TextPreprocessor

model = BGEM3FlagModel("BAAI/bge-m3", use_fp16=True)
```

```python
import chromadb
chroma_client = chromadb.PersistentClient(path="../db")
```

```python
collection = chroma_client.get_or_create_collection(name="neuclir-titles-ru-bge-m3", metadata={"hnsw:space": "cosine"})
```

```python
import pandas as pd
from explore.funcs import load_datasets

datasets = load_datasets(["ru"])
dataset = datasets["ru"]
docs = pd.DataFrame(dataset.docs_iter())
```

```python
docs.head()
```

```python
preprocessor = TextPreprocessor("russian")

docs["title"] = docs["title"].fillna("")
#docs["title"] = docs["title"].apply(preprocessor.preprocess_text)

docs
```

```python


# Split the data into batches
batches = [docs[i:i+100] for i in range(0, len(docs), 100)]

for i, batch in enumerate(tqdm(batches)):
    title_embeddings = model.encode(batch["title"].to_list(), return_dense=True, return_sparse=True, return_colbert_vecs=True)
    metadata = []
    for j in range(len(batch)):
        metadata.append({"url": batch["url"].iloc[j], "text": batch["text"].iloc[j]})

    collection.add(documents=batch["title"].to_list(), embeddings=title_embeddings["dense_vecs"], ids=batch["doc_id"].to_list(), metadatas=metadata)

```
